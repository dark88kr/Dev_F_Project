# 2. 기본 통계 및 전환율 구조 파악

```python
# 전환 여부 파생변수 생성
df["converted"] = df["transactions"].fillna(0).apply(lambda x: 1 if x >= 1 else 0)

# 주요 전환 관련 통계 요약
summary = {
    "전체 세션 수": len(df),
    "전환 세션 수": df["converted"].sum(),
    "비전환 세션 수": (df["converted"] == 0).sum(),
    "전환율 (%)": round(df["converted"].mean() * 100, 2),
    "평균 히트 수": round(df["hits"].mean(), 2),
    "평균 구매 금액": round(df["totalTransactionRevenue"].dropna().mean() / 1_000_000, 2), # 전체 세션 기준 누적 거래 금액
    "평균 페이지뷰 수": round(df["pageviews"].mean(), 2),
    "신규 방문 비율": round((df["newVisits"] == 1).sum() / len(df) *100,2),
    "평균 체류 시간 (초)": round(df["timeOnSite"].mean(), 2),
    "이탈률 (bounces 비율)": round(df["bounces"].fillna(0).mean() * 100, 2)
}

# 결과를 보기 좋게 정리
summary_df = pd.DataFrame.from_dict(summary, orient="index", columns=["값"])
print(summary_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df["totalTransactionRevenue"] = round(df["totalTransactionRevenue"]/1_000_000,2)

# 시각화 대상 컬럼 목록
cols = [
    "hits", "pageviews", "timeOnSite",
    "newVisits", "bounces", "totalTransactionRevenue"
]

for col in cols:
    plt.figure(figsize=(8, 4))

    # 연속형 변수 → 히스토그램 + 평균/중앙값
    if col in ["hits", "pageviews", "timeOnSite", "totalTransactionRevenue"]:
        sns.histplot(data=df, x=col, bins=50, kde=True)

        # 평균 및 중앙값 계산
        mean_val = df[col].mean()
        median_val = df[col].median()

        # 평균선(빨간색), 중앙값선(파란색) 추가
        plt.axvline(mean_val, color='red', linestyle='--', label=f'평균: {mean_val:.2f}')
        plt.axvline(median_val, color='blue', linestyle='-.', label=f'중앙값: {median_val:.2f}')

        plt.title(f"{col} 분포 (히스토그램)")
        plt.legend()

    # 범주형/이진 변수 → 막대그래프 + 비율 표시
    else:
        value_counts = df[col].value_counts(dropna=False).sort_index()
        total = value_counts.sum()
        sns.barplot(x=value_counts.index.astype(str), y=value_counts.values)

        # 막대 위에 텍스트 (개수 + 비율) 표시
        for i, val in enumerate(value_counts.values):
            ratio = (val / total) * 100
            plt.text(i, val + total * 0.01, f'{val} ({ratio:.1f}%)', ha='center', va='bottom')

        plt.title(f"{col} 값 분포 (막대그래프)")
        plt.xlabel(col)
        plt.ylabel("빈도 수")

    plt.tight_layout()
    plt.show()
```

# 3. 변수별 전환율 및 수익률 차이 확인

```python
# 기존 변수 + 추가 변수 포함
extended_categorical_vars = [
    "deviceCategory", "browser", "operatingSystem",
    "channelGrouping", "country",
    "campaign", "productListName", "pageTitle"
]

# 변수별 전환율 비교 분석
conversion_by_category_extended = []
for var in extended_categorical_vars:
    group = df.groupby(var)["converted"].agg(["count", "sum", "mean"]).reset_index()
    group.columns = [var, "세션 수", "전환 수", "전환율 (%)"]
    group["전환율 (%)"] = (group["전환율 (%)"] * 100).round(2)
    group["분류 변수"] = var
    conversion_by_category_extended.append(group)

# 하나의 테이블로 결합
result_extended_df = pd.concat(conversion_by_category_extended, ignore_index=True)
result_extended_df = result_extended_df.sort_values(["분류 변수", "전환율 (%)"], ascending=[True, False])

# 변수별 전환율 요약 결과 출력
for var in extended_categorical_vars:
    print(f"\n📌 변수: {var}")
    display(result_extended_df[result_extended_df["분류 변수"] == var])

```

```python
# 변수별 전환율 요약 결과 출력 (NaN 제거 + 해당 변수만 출력)
for var in extended_categorical_vars:
    print(f"\n📌 변수: {var}")
    display_cols = [var, "세션 수", "전환 수", "전환율 (%)"]
    display_df = result_extended_df[result_extended_df["분류 변수"] == var][display_cols]
    print(display_df.to_string(index=False))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 시각화할 변수 목록 (2단계에서 분석한 범주형 변수)
categorical_vars_to_plot = [
    "deviceCategory", "browser", "operatingSystem",
    "channelGrouping", "country",
    "campaign", "productListName", "pageTitle"
]

# 시각화 함수
def plot_conversion_rate_by_category(data, var):
    group = data.groupby(var)["converted"].agg(["count", "sum", "mean"]).reset_index()
    group.columns = [var, "세션 수", "전환 수", "전환율"]
    group["전환율"] = group["전환율"] * 100
    group = group.sort_values("전환율", ascending=False).head(10)  # 상위 10개만 표시

    fig, ax1 = plt.subplots(figsize=(10, 5))
    sns.barplot(data=group, x=var, y="전환율", ax=ax1, color="skyblue")
    ax1.set_ylabel("전환율 (%)", color="blue")
    ax1.set_title(f"{var}별 전환율 (상위 10개 카테고리)")

    for label in ax1.get_xticklabels():
        label.set_rotation(45)

    plt.tight_layout()
    plt.show()

# 모든 변수에 대해 시각화 실행
for var in categorical_vars_to_plot:
    plot_conversion_rate_by_category(df, var)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 전환 여부 파생 변수 생성
df["converted"] = df["transactions"].fillna(0).apply(lambda x: 1 if x >= 1 else 0)

# 분석할 변수 목록
category_cols = [
    "deviceCategory", "browser", "operatingSystem", "channelGrouping",
    "country", "campaign", "productListName", "pageTitle"
]

# 변수별 세션 수, 전환 수, 전환율 집계
plot_data = []
for col in category_cols:
    summary = df.groupby(col).agg(
        total_sessions=("fullVisitorId", "count"),
        converted_sessions=("converted", "sum")
    ).reset_index()
    summary["conversion_rate"] = round(summary["converted_sessions"] / summary["total_sessions"] * 100, 2)
    summary["category"] = col
    summary.rename(columns={col: "value"}, inplace=True)
    plot_data.append(summary)

plot_df = pd.concat(plot_data, ignore_index=True)

# 전환율 기준 상위 10개 카테고리만 필터링
top10_df = (
    plot_df.groupby("category", group_keys=False)
    .apply(lambda x: x.sort_values("conversion_rate", ascending=False).head(10))
)

# 시각화
import matplotlib.ticker as mtick
import matplotlib.pyplot as plt
import seaborn as sns
import math

sns.set(style="whitegrid")
num_vars = len(category_cols)
fig, axes = plt.subplots(num_vars, 1, figsize=(10, 5 * num_vars))

for idx, category in enumerate(category_cols):
    data = top10_df[top10_df["category"] == category]
    ax = axes[idx]
    sns.barplot(data=data, y="value", x="conversion_rate", ax=ax, palette="viridis")
    ax.set_title(f"{category}별 전환율 상위 10개", fontsize=14)
    ax.set_xlabel("전환율 (%)")
    ax.set_ylabel(category)

    # 막대 위에 전환율, 세션 수, 전환 수 표시
    for i, row in data.iterrows():
        ax.text(row["conversion_rate"] + 0.1, i % 10,
                f'{row["conversion_rate"]:.2f}%\n({row["converted_sessions"]}/{row["total_sessions"]})',
                color='black', va='center')

plt.tight_layout()
plt.show()
```

# 4. 이상 패턴 탐색

```python
# 분석 대상 컬럼
behavior_cols = ["hits","timeOnSite", "pageviews", "bounces"]

# 3단계 결과를 저장할 리스트
behavior_summary = []

for col in behavior_cols:
    total_sessions = df.shape[0]
    non_null_sessions = df[col].notnull().sum()
    zero_sessions = (df[col] == 0).sum()
    converted_mean = df[df["converted"] == 1][col].mean()
    non_converted_mean = df[df["converted"] == 0][col].mean()

    behavior_summary.append({
        "컬럼명": col,
        "전체 세션 수": total_sessions,
        "결측치 수": total_sessions - non_null_sessions,
        "0인 값 수": zero_sessions,
        "전환 세션 평균": round(converted_mean, 2),
        "비전환 세션 평균": round(non_converted_mean, 2),
    })

# 결과를 데이터프레임으로 변환
behavior_summary_df = pd.DataFrame(behavior_summary)

# 보기 좋게 요약 출력
print("📊 3단계: 이상 행동 패턴 분석 요약\n")
print(behavior_summary_df.to_string(index=False))
```
